```python
import os #importei biblioteca do sistema operativo, para interagir com ele
import json #importei biblioteca json para (JavaScript Object Notation), para trocar dados entre sistemas

class GestorTareas: #criei uma classe com o nome GestorTareas
    def __init__(self): #usei o metodo init que é o construtor de classe e será chamado quando um objeto for criado e receba o parametro self
        self.tareas = [] #criei uma lista vazia com o nome tareas
        self.cargar_tareas() #metodo para carregar uma tarefa no formato json

    def mostrar_menu(self): #criei meu método para mostrar meu menu de kanban
        print("1. Adiciona sua tarefa Francisco")
        print("2. Marcar a tarefa como completada")
        print("3. Apagar tarefa")
        print("4. Mostrar tarefas a fazer")
        print("5. Sair")
        #ele vai imprimir na tela as opções acima escritas

    def agregar_tarea(self): #metodo para agregar tarefa
        tarea = input("Ingrese la nueva tarea: ") #atributo tarea que sera uma entrada str do usuário
        self.tareas.append({"tarea": tarea, "completada": False}) #tenho o atributo tarea que vai ser igual a aquela tarefa que acabei de escrever e o atributo completada com o valor Booleano False, porque acabou de ser criada
        print("Tarea agregada con éxito!") #mensagem para o usuario

    def marcar_como_completada(self): #metodo para definir se completei ou não uma tarefa
        self.mostrar_tareas() #mostrar as tarefas que guardei
        indice = int(input("Ingrese el número de la tarea completada: ")) - 1 #posso escrever o número daquela tarefa que foi completada
        if 0 <= indice < len(self.tareas): #se o usuario digita um numero dentro das opções o valor da tarefa muda para True
            self.tareas[indice]["completada"] = True #agora a tarefa tem o valor Booleano True, porqu foi finalizada
            print("Tarea marcada como completada.") #mensagem para o usuario
        else: #condicao contraria, vai imprimir que o número nao é valido
            print("Índice de tarea no válido.")

    def eliminar_tarea(self):
        self.mostrar_tareas() #mostra as tarefas que eu escrevi
        indice = int(input("Ingrese el número de la tarea a eliminar: ")) - 1 #pede pro usuario entrar a opção
        if 0 <= indice < len(self.tareas): #o valor inserido deve estar entre os valores que o indice mostra
            tarea_eliminada = self.tareas.pop(indice) #opcao para eliminar a tarefa da lista com o comando pop
            print(f"Tarea eliminada: {tarea_eliminada['tarea']}") #mensagem pro usuario
        else: #caso contrario, mostra outro mensagem pro usuario
            print("Índice de tarea no válido.")

    def mostrar_tareas(self): #verificarmos se a lista esta vazia
        if not self.tareas:
            print("No hay tareas.")
        else: #no caso da lista não estar vazia vamos com o for
            for i, tarea in enumerate(self.tareas, start=1): #enumerar as tarefas que temos guardadas
                estado = "Completada" if tarea["completada"] else "Pendiente" #estado dessas tarefas
                print(f"{i}. {tarea['tarea']} - {estado}") #mostrar pro usuario

    def guardar_tareas(self): #função que vai me permitir guardar minha entrada de dados como um arquivo
        with open("tareas.json", "w") as archivo: #estou falando pro sistema que a funcao é para ser guardada como arquivo json
            json.dump(self.tareas, archivo) #converter a tarefa que adicionei em uma string json

    def cargar_tareas(self):
        if os.path.exists("tareas.json"): #pergunto se esse arquivo existe no sistema operativo
            with open("tareas.json", "r") as archivo: #condição para abrir essa string como um arquivo
                self.tareas = json.load(archivo) #uso o comnado load para cargar esse arquivo json

                # de acordo com o aprendido nas aulas com Pedro e Larissa, a ordem é a seguinte: 1. Devo importar as bibliotecas que vou a usar
                # 2. Vou criar a classe ou superclase
                # Logo depois crio meus metodos que sao funcoes em class
                # Logo depois vou com minha condicao ou funcionamento do codigo em repeticao o loop

    def ejecutar(self): #função que vai executar a função de acordo com o que o usuario esteja precisando
        while True: #se a opção esta no menu fara
            self.mostrar_menu()
            opcion = input("Seleccione una opción: ")

            if opcion == "1":
                self.agregar_tarea()
            elif opcion == "2":
                self.marcar_como_completada()
            elif opcion == "3":
                self.eliminar_tarea()
            elif opcion == "4":
                self.mostrar_tareas()
            elif opcion == "5":
                self.guardar_tareas()
                print("¡Hasta luego!")
                break
            else: #senão vai imprimir para o usuario que opção não é válida
                print("Opción no válida. Por favor, elija nuevamente.")

if __name__ == "__main__": #comum usar em python  Se o módulo estiver sendo importado como um módulo em outro programa, o código dentro do bloco if não será executado
    gestor = GestorTareas()
    gestor.ejecutar()
```
